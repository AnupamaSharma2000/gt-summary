```python
from great_tables import GT
from great_tables.data import gtcars
import pandas as pd

df = pd.DataFrame(gtcars)
print(df.shape)
print(df.dtypes)
print(df.head(3))
```

```python


numeric_cols = df.select_dtypes(include="number").columns.tolist()
print("Numeric columns:", numeric_cols)

GT(df[["mfr", "model", "ctry_origin"] + numeric_cols].head(10))
```

```python
(
    GT(df[["mfr", "model", "ctry_origin"] + numeric_cols].head(10))
    .tab_header(title="Sports Cars", subtitle="Top 10 by price")
    .fmt_number(columns=["hp", "trq", "mpg_c", "mpg_h"], decimals=1)
    .fmt_currency(columns=["msrp"])
)
```

```python
numeric_cols = ["hp", "trq", "mpg_c", "mpg_h", "msrp"]

group_stats = (
    df.groupby("ctry_origin")[numeric_cols]
    .mean()
    .round(2)
    .reset_index()
)
print(group_stats)
print()
print("Shape:", group_stats.shape)
print("Dtypes:\n", group_stats.dtypes)
```

```python
(
    GT(group_stats, groupname_col="ctry_origin")
    .tab_header(title="Avg Stats by Country", subtitle="Mean across all cars")
    .fmt_number(columns=["hp", "trq", "mpg_c", "mpg_h"], decimals=1)
    .fmt_currency(columns=["msrp"])
)
```

```python
bad_shape = (
    df.groupby("ctry_origin")[numeric_cols]
    .mean()
    .round(2)
    # no reset_index() here
)

print(bad_shape)
print()
print("Index:", bad_shape.index.name)  # ctry_origin is the index, not a column

GT(bad_shape)
```

```python
#what happens with a string column in the mix ─────────────────────
group_stats_with_label = group_stats.copy()
group_stats_with_label["label"] = ["DE", "IT", "JP", "UK", "US"]

(
    GT(group_stats_with_label, groupname_col="ctry_origin")
    .grand_summary_rows(
        fns={"Overall Mean": lambda x: round(x.mean(), 2)}
    )
)
```

grand_summary_rows passes the entire DataFrame to your lambda — including string columns. It does not filter to numeric only. If any string column is present, it crashes with TypeError: Cannot perform reduction 'mean' with string dtype.



```python
(
    GT(group_stats, groupname_col="ctry_origin")
    .tab_header(title="Avg Stats by Country", subtitle="Mean across all cars")
    .fmt_number(columns=["hp", "trq", "mpg_c", "mpg_h"], decimals=1)
    .fmt_currency(columns=["msrp"])
    .grand_summary_rows(
        fns={
            "Overall Mean": lambda x: x.mean(numeric_only=True).round(2),
            "Overall Std":  lambda x: x.std(numeric_only=True).round(2),
        }
    )
)
```

```python
#raw data shape fmt_nanoplot needs
# It needs ONE cell = ONE space-separated string of numbers
#build that manually first to understand it

test_data = pd.DataFrame({
    "country": ["Germany", "Italy", "Japan"],
    "values":  ["419 450 380 500 410",   # Germany hp values as a string
                "600 580 650 620 590",   # Italy
                "559 540 580"]           # Japan
})

print(test_data)

GT(test_data, rowname_col="country").fmt_nanoplot(columns="values", plot_type="bar")
```

```python
# real sparkline strings from the actual gtcars data 
# For each country,take the raw hp values from ALL cars (not the mean)
# and join them into a space-separated string

sparklines = (
    df.groupby("ctry_origin")["hp"]
    .apply(lambda vals: " ".join(str(round(v, 1)) for v in vals.dropna()))
    .reset_index()
    .rename(columns={"hp": "hp_dist"})
)

print(sparklines)
```

```python
#merge into group_stats and render the full table

group_stats_spark = group_stats.merge(sparklines, on="ctry_origin")

print(group_stats_spark.columns.tolist())  # confirm hp_dist is there

(
    GT(group_stats_spark, groupname_col="ctry_origin")
    .tab_header(title="Avg Stats by Country", subtitle="Distribution of HP shown")
    .fmt_number(columns=["hp", "trq", "mpg_c", "mpg_h"], decimals=1)
    .fmt_currency(columns=["msrp"])
    .fmt_nanoplot(columns="hp_dist", plot_type="bar")
    .grand_summary_rows(
        fns={"Overall Mean": lambda x: x.mean(numeric_only=True).round(2)}
    )
)
```

```python
#plot_type=line 
(
    GT(group_stats_spark, groupname_col="ctry_origin")
    .tab_header(title="Avg Stats by Country", subtitle="Line distribution")
    .fmt_nanoplot(columns="hp_dist", plot_type="line")
)


```

### What We Now Know the Module Must Do

1. Detect numeric columns via select_dtypes(include="number")
2. If group_by:
     groupby--> mean --> reset_index()         <-- produces the right shape
3. Build sparkline strings per numeric column from the ORIGINAL df
     groupby --> apply(join) --> reset_index()
4. Merge sparkline cols into the stats DataFrame
5. Call GT(stats_df, groupname_col=group_by)
6. Chain .fmt_number(numeric_cols)
7. Chain .fmt_nanoplot(sparkline_cols, plot_type="bar")
8. Chain .grand_summary_rows(fns with numeric_only=True)
9. Return the GT object
